## Deal with Missing Data

1. Drop Data:
    1. Drop the Whole row
    2. Drop the whole column
    
2. Replace Data:
    1. replace mean 
    2. replace mode [freqency]
    3. replace by median
    4. replace if based on the other function

```python
df.isnull().sum()
```

1. normalized-losses    41
2. num-of-doors          2
3. bore                  4
4. stroke                4
5. horsepower            2
6. peak-rpm              2
7. price                 4

```python
df['normalized-losses'] = df['normalized-losses'].astype('float64')
# df['num-of-doors'] = df['num-of-doors'].astype('float64').mean(axis = 0)
df['bore'] = df['bore'].astype('float64')
df['stroke'] = df['stroke'].astype('float64')
df['horsepower'] = df['horsepower'].astype('float64')
df['peak-rpm'] = df['peak-rpm'].astype('float64')
df['price'] = df['price'].astype('float64')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['normalized-losses','num-of-doors','drive-wheels','engine-location','fuel-system','engine-type'], axis= 1)
```

```python
# df.info()
```

```python
df = df.drop(['make','fuel-type','aspiration','body-style'],axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
df['num-of-cylinders']
```

```python
df['num-of-cylinders'].unique()
```

```python
df['num-of-cylinders'] = df['num-of-cylinders'].replace({'one':1, 'two':2,'three':3, 'four':4,'five':5, 'six':6,'seven':7, 
                                                          'eight':8, 'nine':9, 'ten':10, 'eleven':11, 'twelve':12})
```

```python
df['horsepower'].unique()
```

```python
df.isnull().sum()
```

### Missing Data Handled [fill using Mean value]

```python
df['bore'] = df['bore'].fillna(df['bore'].mean())
df['stroke'] = df['stroke'].fillna(df['stroke'].mean())
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
df['peak-rpm'] = df['peak-rpm'].fillna(df['peak-rpm'].mean())
df['price'] = df['price'].fillna(df['price'].mean())
```

### Indepdendent variable[x] and Dependent Variable [y]

```python
# Independent Variable
x = df.drop('price',axis=1)
```

```python
x.head()
```

```python
#Dependent Variable
y = df['price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
# spliting data into training and test (i.e. 80% - train & 20% - test)
```

### The data is non-normalize

```python
linear_regression = LinearRegression()
```

```python
print(linear_regression)
```

*The data training is finished.*

**Now we need to test the data:**

```python
# we predict by using the x_test data

y_pred = linear_regression.predict(x_test)
```
