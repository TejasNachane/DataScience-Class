# Random Permutations

1. A permutation refers to an arrangement of elements Example [3,2,1] is permutation of [1,2,3] and vice versa
2. In Numpy Random module provides two methods:
    1. shuffle():
        1. Shuffle means changing arrangement of elements
        2. Shuffle method makes changes to the original array
    2. permutation():
        1. Generate random permutation of elements in array
        2. permetuation method return re-arrange array (and leaves the original array un-changes)

```python
from numpy import random
import numpy as np

my_array = np.array([10,20,30,40,50])
random.shuffle(my_array)
print(my_array)
```

```python
# shuffle not use third variable.
from numpy import random
import numpy as np

my_array = np.array([10,20,30,40,50])
new_array = random.shuffle(my_array)
print(new_array)
```

```python
my_array = np.array([10,20,30,40,50])

new_array = random.permutation(my_array)
print(new_array)
```

# Visulize Distribution with Seaborn

1. DistPlot:

2. Importing:
    1. Matplotlib:
        1. import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.distplot([0,10,20,30,40,50,60,70])
plt.show()
```
